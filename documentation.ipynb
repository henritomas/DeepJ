### Piano2Vec (midi-similarity.py)


*   Reduces Piano MIDI pitches into its harmonics (Harmonic Reduction)
*   Embeds harmonics using Word2Vec model trained on 140 MIDI files from the MAESTRO dataset
* Converts piano MIDI to vector using this trained model, and compare with others using Cosine Similarity 
* Expected Results: Generated MIDIs of Mood 1 should have higher Cosine Similarity to other generated MIDIs of the same mood relative to generated MIDIs of different moods.
